```python
import numpy as np


def check_if_dominant(arr: np.array, data: np.array):
	comp_arr = []
	for el in data:
		if np.array_equal(arr, el): continue
		comp = arr > el
		comp_arr.append(sum(comp))

	comp_arr = np.array(comp_arr)
	print(comp_arr)
	return sum(comp_arr == 0)


def generate_ans(data: np.array):
	return [0 if check_if_dominant(arr, data) else 1 for arr in data]


rand = np.random.RandomState(3)
X = rand.randint(0, 10, (100, 5))
print(generate_ans(X))
```

```python
x = np.array([2,3])
data =  np.array([[1,5],[2,3],[10,1],[1,1]])
check_if_dominant(x,data)              
```
